```python
df['2022 Population'].unique()
df['2022 Population'] = pd.to_numeric(df['2022 Population'], errors='coerce')
```

```python
# corr and mean will work only on numeric values
# plot population, so select population columns in ascending order to see the trend of growth
df1=df.groupby('Continent')[df.columns[5:13][::-1]].mean(numeric_only=True).sort_values(by='2022 Population',ascending=False)
```

```python
df1.columns
df2=df1.transpose()
#df1.columns[1:9][::-1]
```

```python
df2.plot() # population versus continent
df1.plot() # continet versus population
```

```python
# box plot to identify outliers
df2.boxplot(figsize=(20,7))
```
